```python
df_patients['birthDate'] = pd.to_datetime(df_patients['birthDate'])

# Calculate age
now = pd.Timestamp('now')

# for i in df_patients['birthDate']:

df_patients['age'] = (now - df_patients['birthDate']).dt.days // 365

# Categorize patients into age groups
def categorize_age(age):
    if age <= 12:
        return 'Child'
    elif 13 <= age <= 19:
        return 'Teen'
    elif 20 <= age <= 59:
        return 'Adult'
    else:
        return 'Senior'


df_patients['ageGroup'] = df_patients['age'].apply(categorize_age)

adult_count = df_patients['ageGroup'].value_counts()['Adult']

print(adult_count)
```
